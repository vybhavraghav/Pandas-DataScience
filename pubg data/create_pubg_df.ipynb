```python
import pandas as pd

weapons= pd.read_html('https://www.eurogamer.net/articles/2019-03-20-pubg-weapon-damage-stats-best-weapons-5414')
```

```python
import requests as r
from bs4 import BeautifulSoup as bs
url=r'https://www.eurogamer.net/articles/2019-03-20-pubg-weapon-damage-stats-best-weapons-5414'
pubg_data=r.get(url)

soup=bs(pubg_data.content,'html.parser')
results=soup.find_all('strong',string=lambda text:'damage stats:' in text.lower())
```

```python
soup.find
weapon_type=[]
for w_t in results:
    group= w_t.text.strip('</strong>')[0:-14]
    
    weapon_type.append(group)
    
```

```python
def create_pubg():
    pubg=pd.DataFrame()
    for i,w_type in enumerate(weapons):
        w_type['Weapon type']=weapon_type[i]
        pubg=pubg.append(w_type,ignore_index=True)
#     pubg.sort_values('Weapon',inplace=True,ignore_index=True)
    pubg['Weapon'][41]='Cowbar'
    pubg['Weapons']=pubg['Weapon'].str.lower()
    return pubg
pubg_=create_pubg()
pubg_
```
