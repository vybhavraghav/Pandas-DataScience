```python
import pandas as pd
import numpy as np
df= pd.read_excel('pubg_main.xlsx')
df.to_csv('pubg_main.csv')
df.to_csv('pubg_main.txt',sep='\t')
df.head()
```

```python
df['Mag Size']=df['Mag Size'].astype('str')
df['Extended Ammo']=0
for i,cell in enumerate(df['Mag Size']):
    if cell.endswith(')'):
        ext=cell.split('(')
        df['Extended Ammo'][i]=ext[1].strip('Ext:)')
        df['Mag Size'][i]=ext[0]
    if cell=='nan':
        df['Mag Size'][i]=np.nan

    
df.head()
```

```python
df['Ammo']=df['Ammo'].astype('str')
df['Ammo']=df['Ammo'].apply(lambda x: x.strip('mmG'))
df['Reload']=df['Reload'].astype('str')
df['Reload']=df['Reload'].apply(lambda x: x.strip('s'))
df.rename(columns={'Ammo':'Ammo (mm)','Reload':'Reload (s)'},inplace=True)

df.head()
```

```python
df.loc[df['Ammo (mm)']=='nan',['Ammo (mm)','Reload (s)']]=np.nan
df[['Mag Size','Reload (s)','Extended Ammo']]=df[['Mag Size','Reload (s)','Extended Ammo']].astype('float')
def ext_mag():
    for i,cell in enumerate(df['Extended Ammo']):
        if cell!=0:
            df['Extended Ammo'][i]=df['Extended Ammo'][i]-df['Mag Size'][i]
    l=df.loc[df['Mag Size'].isna()==True].index
    df['Extended Ammo'][l]=np.nan
            
ext_mag()
```
