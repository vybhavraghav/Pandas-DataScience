```python
import pandas as pd
df = pd.read_csv('google play store apps/googleplaystore.csv')
df=df.loc[df["Category"]=='GAME']
df=df.loc[~(df['App'].duplicated(keep='last'))]
df.sort_values('App',inplace=True)
df
```

```python
df.to_csv('google play store apps/Playstoregames.csv',index=False)
df.to_csv('google play store apps/Playstoregames.txt',sep='\t',index=False)
df.to_excel('google play store apps/Playstoregames.xlsx',index=False)
df.to_html('google play store apps/Playstoregames.html',index=False)
```
